```python
import urllib
import requests
import json
import math
import pandas as pd

page_size = 100
page = 1
years = ('2016', '2017')
level = "Gminy"

params = {'format': 'jsonapi', 'unit-level': levels[level], 'year': years, 'page': page, 'page-size': page_size}
#unit-leve, 2 województwo itd
id = '2065'
results = []
data = requests.get("https://bdl.stat.gov.pl/api/v1/data/by-variable/"+id+"?", params)
data = json.loads(data.content.decode('utf-8'))
total_pages = data['meta']['totalRecords']
pages = int(math.ceil(total_pages/page_size))
for i in range(pages):
    params['page'] = i
    data = requests.get("https://bdl.stat.gov.pl/api/v1/data/by-variable/"+id+"?", params)
    if data.status_code == 200:
        data = json.loads(data.content.decode('utf-8'))
        for item in data['data']:
            result = {'name': item['attributes']['name'], 'unit_id': item['id']}
            for i in range(len(item['attributes']['values'])):
                result.update({item['attributes']['values'][i]['year']: item['attributes']['values'][i]['val']})
            results.append(result)
```
